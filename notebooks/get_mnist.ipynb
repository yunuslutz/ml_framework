```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
data = datasets.fetch_openml(
'mnist_784',
version=1,
return_X_y=True)
```

```python
pixel_values, targets = data
targets = targets.astype(int)
```

```python
df = pd.DataFrame(np.column_stack((pixel_values, targets)))
```

```python
df.to_csv('../input/train.csv', index=False)
```
